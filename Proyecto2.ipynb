```python
import pandas as pd
url= "https://raw.githubusercontent.com/MiguelBarranco345/Tareas_de_THC/refs/heads/main/ECOVSCOVID.csv"
df=pd.read_csv(url)
resume = df.describe()
print(resume)
```

```python
columnas_faltantes = df.isnull().any()
print(columnas_faltantes)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df_mexico = df[df['COUNTRY'] == 'Mexico']
print(df_mexico)
```

```python
ResumenMexico=df_mexico.describe()
print(ResumenMexico)
```

```python
df_mexico = df[df['COUNTRY'] == 'Mexico'].set_index('DATE')
tc = df_mexico['TC'][0:-1:7]
td = df_mexico['TD'][0:-1:7]
Dates = df_mexico.index[0:-1:7]

plt.figure(figsize=(12, 6))
plt.plot(Dates, tc, 'o:r', ms=4, label='Enfermos')
plt.bar(Dates, tc, color='red', alpha=0.2)
plt.plot(Dates, td, 'o-b', ms=4, label='Decesos')
plt.bar(Dates, td, color='blue', alpha=0.2)

plt.xticks(ticks=Dates,rotation=90,fontsize=10)
plt.grid(alpha=0.5)
plt.title('Casos confirmados y decesos en México (Cada 7 días)')
plt.ylabel('Casos confirmados')
plt.legend()
plt.show()
```

```python

Poblacion = df_mexico['POP'][0:-1:7]
Dates = df_mexico.index[0:-1:7]

plt.figure(figsize=(12, 6))
plt.bar(Dates, tc, color='red', alpha=0.2)
plt.plot(Dates, Poblacion, 'o-b', ms=4, label='Poblacion')
plt.bar(Dates, Poblacion, color='blue', alpha=0.2)

plt.xticks(ticks=Dates,rotation=90,fontsize=10)
plt.grid(alpha=0.5)
plt.title('Poblacion en Mexico durante el periodo COVID')
plt.ylabel('Poblacion')
plt.legend()
plt.show()
```

```python

max_sti = df['STI'].max()
min_sti = df['STI'].min()
max_sti_countri = df[df['STI'] == max_sti]
min_sti_countri = df[df['STI'] == min_sti]
print(max_sti_countri[['COUNTRY', 'STI','CODE']])
print(min_sti_countri[['COUNTRY', 'STI','CODE']])
```

```python
alto_sti = 3
bajo_sti = 1
sti_min = df.groupby('CODE')['STI'].min()
sti_maximos = df.groupby('CODE')['STI'].max()
paises_alto_sti = sti_min[sti_min >= alto_sti]
paises_bajo_sti = sti_maximos[sti_maximos < bajo_sti]

print(paises_bajo_sti)
print(paises_alto_sti)
```

```python
paises = ['ZWE','LSO','UGA', 'ARM', 'JEY','COM']
resultados = {'Pais': [], 'Tasa_Enfermos': [], 'STI': []}

for pais in paises:
    pais_data = df[df['CODE'] == pais]
    total_casos = pais_data['TC'].sum()
    total_poblacion = pais_data['POP'].sum()
    tasa_enfermos = (total_casos / total_poblacion) * 100
    sti = pais_data['STI'].max()
    resultados['Pais'].append(pais)
    resultados['Tasa_Enfermos'].append(tasa_enfermos)
    resultados['STI'].append(sti)

resultados_df = pd.DataFrame(resultados)
print(resultados_df)

plt.figure(figsize=(12, 7))
plt.bar(resultados_df['Pais'], resultados_df['Tasa_Enfermos'], color='g', alpha=0.6, label='Tasa de Enfermos')
plt.plot(resultados_df['Pais'], resultados_df['STI'] * 100, color='orange', marker='o', label='STI (Escalado)')
plt.title('Porcentaje de enfermos vs STI')
plt.xlabel('País')
plt.ylabel('Porcentaje (%)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python


paises = ['AUS', 'SGP', 'ISL', 'BFA', 'MLI', 'AFG']
resultados = {'Pais': [], 'Tasa_Mortalidad': [], 'Indice_Desarrollo': []}

for pais in paises:
    pais_data = df[df['CODE'] == pais]

    #Aqui suma los datos de la población
    total_muertes = pais_data['TD'].sum()
    total_poblacion = pais_data['POP'].sum()
    #Porcentaje de tasa de mortalidad, porque puede haber muy poquitos muertos en un país pero que tenga muy poquita población
    #Por ejemplo que sea una poblacion de 5 y una de 10 y en ambos se mueren 5, tienen el mismo numero de fallecidos pero en uno aún queda el 50% de población
    tasa_mortalidad = (total_muertes / total_poblacion) * 100
    #iloc extrae la información de la fila
    indice_desarrollo = pais_data['HDI'].iloc[-1]
    #se ponen los resultados en los espacios que creamos arriba.
    resultados['Pais'].append(pais)
    resultados['Tasa_Mortalidad'].append(tasa_mortalidad)
    resultados['Indice_Desarrollo'].append(indice_desarrollo)

resultados_df = pd.DataFrame(resultados)
print(resultados_df)

plt.figure(figsize=(10, 6))
plt.bar(resultados_df['Pais'], resultados_df['Tasa_Mortalidad'], color='b', alpha=0.6, label='Tasa de Mortalidad')
plt.plot(resultados_df['Pais'], resultados_df['Indice_Desarrollo'] * 100, color='r', marker='o', label='Índice de Desarrollo Humano')
plt.title('Tasa de Mortalidad vs Índice de Desarrollo Humano')
plt.xlabel('País')
plt.ylabel('Valor')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
paises = ['IRL','NOR','QAT', 'CAF', 'ZWE','SLE']
resultados = {'Pais': [], 'Tasa_Enfermos': [], 'GDPCAP': []}

for pais in paises:
    pais_data = df[df['CODE'] == pais]
    total_casos = pais_data['TC'].sum()
    total_poblacion = pais_data['POP'].sum()
    tasa_enfermos = (total_casos / total_poblacion) * 100
    gdpcap = pais_data['GDPCAP'].iloc[-1]
    resultados['Pais'].append(pais)
    resultados['Tasa_Enfermos'].append(tasa_enfermos)
    resultados['GDPCAP'].append(gdpcap)
resultados_df = pd.DataFrame(resultados)
print(resultados_df)
plt.figure(figsize=(12, 7))
plt.bar(resultados_df['Pais'], resultados_df['Tasa_Enfermos'], color='r', alpha=0.6, label='Tasa de Enfermos')
plt.plot(resultados_df['Pais'], resultados_df['GDPCAP'] * 100, color='blue', marker='o', label='GDPCAP')
plt.title('Porcentaje de enfermos vs GDPCAP')
plt.xlabel('País')
plt.ylabel('Valor')
plt.legend()
plt.tight_layout()
plt.show()
```
